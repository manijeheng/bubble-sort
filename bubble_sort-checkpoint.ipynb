```python
import random
import time
import numpy as np
a = 0
# ایجاد آرایه ده هزار تایی از اعداد تصادفی بین صفر تا صد هزار
array = np.random.randint(0,100000,size=1000)

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                a = arr[j]
                arr[j] = arr[j+1]
                arr[j+1] = a
              
            

# محاسبه زمان اجرای مرتب سازی حبابی
start_time = time.time()
bubble_sort(array)
end_time = time.time()

print(f"زمان اجرای مرتب سازی حبابی: {end_time - start_time} ثانیه")
```
